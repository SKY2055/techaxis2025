```python
#Student_Database
"""Given the following complex dictionary representing a student database:

school_data = {
    "class_10A": {
        "students": [
            {"name": "John", "grades": {"math": 80, "science": 90}},
            {"name": "Sarah", "grades": {"math": 95, "science": 88}}
        ]
    }
}



output:
John: maths:80, science:90
Sarah: maths:95, science:88
"""

school_data = {
    "class_10A": {
        "students": [
            {"name": "John", "grades": {"math": 80, "science": 90}},
            {"name": "Sarah", "grades": {"math": 95, "science": 88}}
        ]
    }
}
for student in school_data["class_10A"]["students"]:
    name= student["name"]
    math= student["grades"]["math"]
    science= student["grades"]["science"]
    print(f"{name}: maths:{math}, science:{science}")
```

```python
#The Data Parser(String Slicing)
"""You have a date in a specific format: date_string = "2023-12-25".

Use  (using square brackets []) to extract the year, month, and day into separate variables.
year should be "2023"
month should be "12"
day should be "25"
 Print them in US format: 12/25/2023 using an f-string."""

date_string = "2023-12-25"
year= date_string[0:4]
month=date_string[6:7]
day=date_string[8:10]

print(year)
print(month)
print(day)

print(f"{month}/{day}/{year}")
```
